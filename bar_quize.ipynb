```python
import matplotlib.pyplot as plt

import numpy as np

x=np.array([45,64,54,22])

y=np.array(['CSE','Mech','ECE','Civil'])

plt.pie(x, labels = y)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

import numpy as np

x=np.array([45,64,54,22])

y=np.array(['python','Maths','English','chemistry'])
myexplode=[0.2,0,0,0]
plt.pie(x, labels = y, explode = myexplode)
plt.show() 
```
