#bar graph

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array(['clad','it','eng','cprograme','single','several','mech','phy','vdc'])
ypoints = np.array([90,75,76,100,60,65,60,100,55])
plt.bar(xpoints, ypoints,color='green',width=0.5) #bar
plt.show()
plt.barh(xpoints,ypoints,color='c',height=0.5) 
plt.show()
plt.bar(xpoints, ypoints,color='black')
plt.show()
plt.bar(xpoints, ypoints,color='hotpink')
plt.show()
plt.barh(xpoints,ypoints,color='blue',height=0.5) 
plt.show()
```

#Pie Charts

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([30, 22, 27, 18])

plt.pie(x)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["cars", "Bikes", "buses", "vans"]
myexplode = [0.2, 0, 0, 0]
mycolors = ["black", "hotpink", "b"]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```
