```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array(['clad','it','eng','cprograme','single','several','mech','phy','vdc'])
ypoints = np.array([90,75,76,100,60,65,60,100,55])

plt.plot(xpoints, ypoints)
plt.show()
plt.bar(xpoints, ypoints,color='green',width=0.5)
plt.show()
plt.scatter(xpoints, ypoints)
plt.show()
plt.pie(ypoints)
plt.show()

```
