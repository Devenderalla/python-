#Markers

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([90,75,76,100,60,65,60,100,55])

plt.plot(ypoints, marker = 'o')
plt.show()
plt.plot(ypoints, 'o:r')
plt.show()
plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

#Line graph

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array(['clad','it','eng','cprograme','single','several','mech','phy','vdc'])
ypoints = np.array([90,75,76,100,60,65,60,100,55])
plt.plot(ypoints, linestyle = 'dotted')
plt.show()
plt.plot(ypoints, linestyle = 'dashed')
plt.show()
plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

#Scatter

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array(['clad','it','eng','cprograme','single','several','mech','phy','vdc'])
ypoints = np.array([90,75,76,100,60,65,60,100,55])
plt.scatter(xpoints, ypoints)  
plt.show()
```

#bar graph

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array(['clad','it','eng','cprograme','single','several','mech','phy','vdc'])
ypoints = np.array([90,75,76,100,60,65,60,100,55])
plt.bar(xpoints, ypoints,color='green',width=0.5) #bar
plt.show()
plt.barh(xpoints,ypoints,color='c',height=0.5) 
plt.show()
```

#Pie Charts

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([30, 22, 27, 18])

plt.pie(x)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["cars", "Bikes", "buses", "vans"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```
